```python
from matplotlib import axis
import pandas as pd
larry = pd.read_csv('/content/Car.csv')
print("Shape of the dataset:", larry.shape)
larry.head(30)
```

```python
#DATA CLEANING
print(larry.isnull().sum())
larry = larry.dropna()
larry = larry.drop_duplicates()

```

```python
#Visualzation
import pandas as pd
import matplotlib.pyplot as plt
larry['Price_in_thousands'] = pd.to_numeric(larry['Price_in_thousands'], errors='coerce')
larry['sales'] = pd.to_numeric(larry['sales'], errors='coerce')
larry = larry.dropna()

larry['Price_Range'] = pd.qcut(larry['Price_in_thousands'], q=4)
sales_by_price = larry.groupby('Price_Range')['sales'].sum()
larry['Fuel_efficiency'] = pd.to_numeric(larry['Fuel_efficiency'], errors='coerce')
larry['Fuel_Category'] = pd.qcut(larry['Fuel_efficiency'], q=4)
sales_by_fuel = larry.groupby('Fuel_Category')['sales'].sum()


plt.figure()
plt.bar(sales_by_price.index.astype(str), sales_by_price.values)
plt.title("Total Sales by Price Range")
plt.xlabel("Price Range (Thousands)")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()

plt.figure()
plt.bar(sales_by_fuel.index.astype(str), sales_by_fuel.values, color='green')
plt.title("Total Sales by Fuel Efficiency Range")
plt.ylabel("Total sales")
plt.xlabel("Fuel Efficiency")
plt.xticks(rotation=45)
plt.show()

plt.figure()
plt.scatter(larry['Horsepower'], larry['Fuel_efficiency'], color = 'red')
plt.title("Horsepower vs Fuel Efficiency")
plt.xlabel("Horsepower")
plt.ylabel("Fuel Efficiency")
plt.show()

```

```python
columns = ['Price_in_thousands', 'Engine_size',
           'Horsepower', 'Fuel_efficiency', 'sales']

for col in columns:
    print("\nColumn:", col)
    print("Mean:", larry[col].mean())
    print("Mode:", larry[col].mode().values)
    print("Variance:", larry[col].var())

```
